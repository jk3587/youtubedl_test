```python
import youtube_dl #pip install 
import webvtt
```

```python
url = 'https://www.youtube.com/watch?v=NiKtZgImdlY'

ydl_opts = {
    'format': 'bestaudio/best',
    #'outtmpl': '%(title)s.%(etx)s',
    'writesubtitles': True,
    'writeautomaticsub': True,
    'quiet': False
}

with youtube_dl.YoutubeDL(ydl_opts) as ydl:
    ydl.download([url])  # Download into the current working directory
```

```python
for caption in webvtt.read('The danger of silence _ Clint Smith-NiKtZgImdlY.en.vtt'):
    print(caption.start)
    print(caption.end)
    print(caption.text)
```
